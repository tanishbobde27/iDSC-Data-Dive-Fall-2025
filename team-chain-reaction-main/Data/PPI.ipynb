```python
def filterDate_PPI(year):
    data = []
    for i in range(len(df_PPI)):
        n = year
        num = str(n)
        iteration = str(df_PPI["observation_date"][i])
        if num in iteration:
            a = df_PPI["observation_date"][i]
            b = df_PPI["ProducerPriceIndex"][i]
            d = {"observation_date" : a, "ProducerPriceIndex" : b}
            data.append(d)

    df = pd.DataFrame(data)   
    return df 
```

```python
df_PPI = pd.read_csv("producerPI.csv")
df_PPI.rename(columns={'PCU33443344': 'ProducerPriceIndex'}, inplace=True) 

df_new = filterDate_PPI(2010)
num = 2011
for i in range(15):
    df2 = filterDate_PPI(num)
    df_new = pd.concat([df_new, df2], ignore_index = True)
    num += 1

df_PPI = df_new
df_PPI
```
