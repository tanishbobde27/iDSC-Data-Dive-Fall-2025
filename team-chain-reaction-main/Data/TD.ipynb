```python
import pandas as pd
import plotly.express as p
import networkx as nx
import geopandas as gpd
import matplotlib.pyplot as plt
```

```python
try:
    df_TD = pd.read_csv("Trade_Data.csv", encoding='latin1')
except UnicodeDecodeError:
    try:
        df_TD = pd.read_csv("Trade_Data.csv", encoding='cp1252')
    except UnicodeDecodeError:
        df_TD = pd.read_csv("Trade_Data.csv", encoding='utf-8-sig')

cols_to_keep = [
    "refYear", "refMonth",
     "reporterISO", "cmdCode", 
     "cifvalue", "fobvalue", "primaryValue",
     "partnerISO", "isGrossWgtEstimated" ]

df_TD = df_TD[cols_to_keep]

index = []
for i in range(len(df_TD)):
    index.append(i)

df_TD.index = index
df_TD
```

```python
df_TD["count"] = 1

d = df_TD.groupby("reporterISO").agg("sum", numeric_only = True).reset_index()
```

```python
fig = px.choropleth(df_TD,
                    locations="reporterISO",
                    color="cifvalue",
                    hover_name="reporterISO",
                    animation_frame="refYear",
                    title="Average CIF Export Value by Country Over Time",
                    color_continuous_scale="Viridis")
fig.show()
```

```python
df_TD["time"] = df_TD["refYear"].astype(str) + "-" + df_TD["refMonth"].astype(str)
fig = p.choropleth(df_TD,
                    locations="reporterISO",
                    color="cifvalue",
                    animation_frame="time",
                    title="Monthly Import Price Changes Over Time",
                    color_continuous_scale="Plasma")
fig.show()
```
