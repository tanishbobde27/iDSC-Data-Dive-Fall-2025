```python
def filterDate_cpi(year):
    data = []
    for i in range(len(df_cpi)):
        n = year
        num = str(n)
        if num in df_cpi["observation_date"][i]:
            a = df_cpi["observation_date"][i]
            b = df_cpi["ConsumerPriceIndex"][i]
            d = {"observation_date" : a, "ConsumerPriceIndex" : b}
            data.append(d)

    df = pd.DataFrame(data)   
    return df 
```

```python
def filterDate_GSCPI(year):
    data = []
    for i in range(len(df_gscpi)):
        n = year
        num = str(n)
        iteration = str(df_gscpi["Date"][i])
        if num in iteration:
            a = df_gscpi["Date"][i]
            b = df_gscpi["GSCPI"][i]
            d = {"Date" : a, "GSCPI" : b}
            data.append(d)

    df = pd.DataFrame(data)   
    return df 
```

```python
def filterDate_PPI(year):
    data = []
    for i in range(len(df_PPI)):
        n = year
        num = str(n)
        iteration = str(df_PPI["observation_date"][i])
        if num in iteration:
            a = df_PPI["observation_date"][i]
            b = df_PPI["ProducerPriceIndex"][i]
            d = {"observation_date" : a, "ProducerPriceIndex" : b}
            data.append(d)

    df = pd.DataFrame(data)   
    return df 
```
