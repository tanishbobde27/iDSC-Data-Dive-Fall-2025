```python
import pandas as pd
```

```python
def filterDate_GSCPI(year):
    data = []
    for i in range(len(df_gscpi)):
        n = year
        num = str(n)
        iteration = str(df_gscpi["Date"][i])
        if num in iteration:
            a = df_gscpi["Date"][i]
            b = df_gscpi["GSCPI"][i]
            d = {"Date" : a, "GSCPI" : b}
            data.append(d)

    df = pd.DataFrame(data)   
    return df 
```

```python
df_1= pd.read_csv("gscpi_data 2.csv")
columns_to_drop = ["Unnamed: 2", "Unnamed: 3", "Unnamed: 4", "Unnamed: 5", "Unnamed: 6", "Unnamed: 7", "Unnamed: 8", "Unnamed: 9", "Unnamed: 10"]
df_globalsupplychain = df_1.drop(columns_to_drop, axis = 1)
df_gscpi = df_globalsupplychain.dropna()
df_gscpi.reset_index(drop = True, inplace = True)

df_new = filterDate_GSCPI(2010)
num = 2011
for i in range(15):
    df2 = filterDate_GSCPI(num)
    df_new = pd.concat([df_new, df2], ignore_index = True)
    num += 1

df_gscpi = df_new
df_gscpi
```

```python
df_GSCPI_FirstThird = filterDate_GSCPI(2010)
num = 2011
for i in range(5):
    df3 = filterDate_GSCPI(num)
    df_GSCPI_FirstThird = pd.concat([df_GSCPI_FirstThird, df3], ignore_index = True)
    num += 1

df_GSCPI_SecondThird = filterDate_GSCPI(2016)
num = 2017
for i in range(4):
    df4 = filterDate_GSCPI(num)
    df_GSCPI_SecondThird = pd.concat([df_GSCPI_SecondThird, df4], ignore_index = True)
    num += 1

df_GSCPI_LastThird = filterDate_GSCPI(2021)
num = 2022
for i in range(4):
    df5 = filterDate_GSCPI(num)
    df_GSCPI_LastThird = pd.concat([df_GSCPI_LastThird, df5], ignore_index = True)
    num += 1
```

```python
df_GSCPI_SecondHalf = filterDate_GSCPI(2018)
num = 2019
for i in range(7):
    df4 = filterDate_GSCPI(num)
    df_GSCPI_SecondHalf = pd.concat([df_GSCPI_SecondHalf, df4], ignore_index = True)
    num += 1
```
